```python
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period = 'max')
```

```python
sp500
```

```python
sp500.index
```

```python
sp500.plot.line(y = 'Close', use_index=True);
```

```python
del sp500['Dividends']
del sp500['Stock Splits']
```

```python
sp500['Tomorrow']  = sp500['Close'].shift(-1) 
#creates a new column where each row holds the next day's closing price(what we are trying to predict)
```

```python
sp500
```

```python
sp500["Target"] = (sp500['Tomorrow'] > sp500['Close']).astype(int)
#creates a new 'Target' column that returns int 1 if the stock price will go up the next day and 0 otherwise
```

```python
sp500
```

```python
sp500 = sp500.loc["1990-01-01":].copy()
# filters the dataset to only include data from 1990-01-01 to the present to achieve more accurate results
# if you modify the filtered result (sp500.loc["1990-01-01":]) directly, it could potentially change the original 
# dataframe unintentionally. So to avoid this I used .copy() to create an entirely seperate object in memory
```

```python
sp500
```

```python

from sklearn.ensemble import RandomForestClassifier 

#RandomForestClassifier is a model for machine learning. They train a bunch of individual decision trees with randomized parameters
#and then average those results together. Therefore they are pretty resistant to overfitting(when the model "memorizes" 
#the training data instead of learning the underlying patterns, leading to inaccurate predictions on new data). 
#This model does not eliminate overfitting completely, but it is harder for it to overfit than it is for other models.
#RandomForestClassifier can also pick up non-linear relationships like with the the "Target" and "Open" columns.
#Obviously most stock prices have a non-linear trend, therefore this model works well for stock price prediction.

model = RandomForestClassifier(n_estimators = 100, min_samples_split = 100, random_state=1)
#n_estimators = number of indivudal decision trees
#min_sample_splits = protects against overfitting by preventing to tree to build very deeply. The higher we set it the lower the accuracy
#random_state = if we run the same model twice, the random numbers that are generated will be in a predictable
#sequence each time by using the seed 1

train = sp500.iloc[:-100] #all of the rows in the dataset except the last 100
test = sp500.iloc[-100:] #the last 100 rows in the dataset
#cannot use cross validation because this is time-series data

predictors = ['Open', 'Close', 'High', 'Low', 'Volume'] #only used the columns that do not have information about the future
model.fit(train[predictors], train['Target']) #The model learns the relationship between the features (train[predictors]) and the target (train['Target'])
```

```python
from sklearn.metrics import precision_score
#Measures how accurate our model is. When we said the market would go up('Target' column) did it actually go up

preds = model.predict(test[predictors])
#Generates the predictions(0,1) if the price will go up or down
```

```python
import pandas as pd

preds = pd.Series(preds, index=test.index)
#Converts array to series. Allows elements of different types to be in the same series
```

```python
preds
```

```python
precision_score(test['Target'], preds)
```

```python
combined = pd.concat([test['Target'], preds], axis=1)
#Combines the data and concatenates along the y-axis(columns)

combined.plot()
#orange line = predictions
#blue line = what actually happened
```

```python
#Iâ€™m going to create a backtesting model that lets me test performance across multiple years, 
#rather than just the last 100 days like I did in the previous model. This is important because, 
#in the real world future, I need to see how the model handles different market conditions and situations 
#over a longer timeframe.

def predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target']) #The model learns the relationship between the features (train[predictors]) and the target (train['Target'])
    preds = model.predict(test[predictors]) #The model applies what it learned during training to predict whether the target is 0 or 1 for each test example.
    preds = pd.Series(preds, index=test.index, name='Predictions') #Converts the predictions array to a series
    combined = pd.concat([test['Target'], preds], axis=1) #Combines the data and concatenates along the y-axis(columns)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    #Every trading year has about 250 days so this takes in 10 years of data(start=2500) and trains the model for 
    #about a year(step=250) and then moving on to the next. So this model will use the first 10 years to predict the 
    #11th year, and then the first 11 years to predict the 12th year and so on.
    
    all_predictions = [] #List of dataframes, where each dataframe is the predictions for a single year

    for i in range(start, data.shape[0], step): #creates preidictions for all the years except the first 10 years or so
        train = data.iloc[0:i].copy() #all of the years prior to the current year
        test = data.iloc[i:(i+step)].copy() #the current year
        predictions = predict(train, test, predictors, model) #call the predict function on the new train/test data and store the data in 'predictions'
        all_predictions.append(predictions) #store 'predictions' value in the all_predictions list

    return pd.concat(all_predictions) #returns a single combined dataframe from the list of dataframes 
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
#Now we will evaluate the error of our predictions

predictions['Predictions'].value_counts() #returns how many days we predicted that the market will go up or down
```

```python
precision_score(predictions['Target'], predictions['Predictions']) #This gives the proportion of "up" predictions that 
# were correct. It shows how well your model performs specifically when predicting the market will go up.
```

```python
predictions['Target'].value_counts() / predictions.shape[0] #This gives a benchmark statistic based on what actually occured 
# by calculating the proportion of rows in the dataset where the market went up or down.
```

```python
# Now we are going to add additional indicators to account for the business cycle, enabling our model 
#to make more accurate future market predictions.

horizons = [2, 5, 60, 250, 1000] # Want to look at rolling means by calculating the mean trading price for the last 2 days, 
# last trading week, last 3 months, last year, and last 4 years. And then I will look at the ratio between today's closing 
# price and the closing price in those periods. This will hint to us that if the market has been going up a ton then it may
# be due for a downturn and vise versa.

new_predictors = [] #will hold the new columns we will create

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean() #calculates the rolling mean for each horizon

    ratio_column = f"Close_ratio_{horizon}" #creates a column named "Close_ratio_(the current horizon)"
    sp500[ratio_column] = sp500['Close'] / rolling_averages['Close'] #adding the ratio column and its values into the sp500 dataframe 
    #If the ratio is significantly greater than 1, the price might be overbought (potential for a downturn).
    #If the ratio is significantly less than 1, the price might be oversold (potential for an upturn).

    trend_column = f"Trend_{horizon}" #creates a column named "Trend_(the current horizon)" which holds the number 
    #of days in the horizon that the stock price actually went up
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()['Target'] #counts how many times the stock price increased 
    #(Target = 1) over the past horizon days using a rolling window 

    new_predictors += [ratio_column, trend_column]
    
# The ratio column will include the previous and current day for the horizon window. The trend column does only includes the 
# previous days which is why some columns have NaN values.
```

```python
sp500
```

```python
sp500 = sp500.dropna() #this will drop the rows that have NaN values so now our new dataframe will start approximately 
#four years after 1990
sp500
```

```python
#Now we will improve our model by changing some of our parameters

model = RandomForestClassifier(n_estimators = 200, min_samples_split = 50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target']) 
    preds = model.predict_proba(test[predictors])[:, 1] #Instead of outputting a 0 or 1 like before, this will output 
    #list of probabilities where each value indicates the likelihood of the instance belonging to class 1
    preds[preds >= 0.6] = 1 #If the predicted probability for class 1 is greater than or equal to 0.6, itâ€™s classified as 1 
    preds[preds < 0.6] = 0 #If the predicted probability is less than 0.6, itâ€™s classified as 0
    preds = pd.Series(preds, index=test.index, name='Predictions') 
    combined = pd.concat([test['Target'], preds], axis=1) 
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions['Predictions'].value_counts()
#this will result in less 1 counts but it will be more accurate in predicting if the price will go up because of the 0.6 threshold
```

```python
precision_score(predictions['Target'], predictions['Predictions'])
```
